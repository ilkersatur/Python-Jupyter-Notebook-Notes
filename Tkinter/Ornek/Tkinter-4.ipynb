```python
from tkinter import * 
import random


bilgisayar_deger={
    "0":"Tas",
    "1":"Kagit",
    "2":"Makas"}



def reset_game():
    button1["state"]="active"
    button2["state"]="active"
    button3["state"]="active"
    oyuncu_label.config(text="Oyuncu")
    bilgisayar_label.config(text="Bilgisayar")
    sonuc_label.config(text="")
    
    
def button_disable():
    """
    button1["state"]="disable"
    button2["state"]="disable"
    button3["state"]="disable"
    """
    button1.config(state="disabled")
    button2.config(state="disabled")
    button3.config(state="disabled")
    


def Tas():
    
    bilgisayar_tahmin=bilgisayar_deger[str(random.randint(0,2))]
    
    if bilgisayar_tahmin=="Tas":
        sonuc="Berabere!"
    
    elif bilgisayar_tahmin=="Makas":
        sonuc="Siz Yendiniz!"
    
    else:
        sonuc="Bilgisayar Yendi!"
        
    sonuc_label.config(text=sonuc)
    oyuncu_label.config(text="Tas")
    bilgisayar_label.config(text=bilgisayar_tahmin)
    button_disable()
    
    

def Kagit():
    
    bilgisayar_tahmin=bilgisayar_deger[str(random.randint(0,2))]
    
    if bilgisayar_tahmin=="Kagit":
        sonuc="Berabere!"
    
    elif bilgisayar_tahmin=="Tas":
        sonuc="Siz Yendiniz!"
    
    else:
        sonuc="Bilgisayar Yendi!"
        
    sonuc_label.config(text=sonuc)
    oyuncu_label.config(text="Kagit")
    bilgisayar_label.config(text=bilgisayar_tahmin)
    button_disable()
    


def Makas():
    
    bilgisayar_tahmin=bilgisayar_deger[str(random.randint(0,2))]
    
    if bilgisayar_tahmin=="Makas":
        sonuc="Berabere!"
    
    elif bilgisayar_tahmin=="Tas":
        sonuc="Bilgisayar Yendi!"
    
    else:
        sonuc="Siz Yendiniz!"
        
    sonuc_label.config(text=sonuc)
    oyuncu_label.config(text="Makas")
    bilgisayar_label.config(text=bilgisayar_tahmin)
    button_disable()
    
    

root=Tk()
root.geometry("400x400")
root.resizable(0,0)

root.title("Tas, Kagit, Makas Oyunu")



Label(root,text="Tas-Kagit-Makas", font=("Calibri",30,"bold"), fg="tomato4").pack(pady=20)

frame=Frame(root)
frame.pack()


oyuncu_label=Label(frame,text="Oyuncu ",font=("Calibri",20,"bold"))#l1
oyuncu_label.pack(side=LEFT)

vs_label=Label(frame,text="VS", font=("Calibri",20,"bold")) #l2
vs_label.pack(side=LEFT)

bilgisayar_label=Label(frame,text=" Bilgisayar", font=("Calibri",20,"bold")) #l3
bilgisayar_label.pack()


sonuc_label=Label(root, text="",font=("Calibri",20,"bold"), width=15, borderwidth=2,relief="solid",bg="white")#l4
sonuc_label.pack(pady=20)

frame2=Frame(root)
frame2.pack()

button1=Button(frame2,text="TAS",  font=("Calibri",20,"bold"), width=7,command=Tas, relief=RAISED)
button2=Button(frame2,text="KAGIT",  font=("Calibri",20,"bold"), width=7,command=Kagit, relief=RAISED)
button3=Button(frame2,text="MAKAS",  font=("Calibri",20,"bold"), width=7,command=Makas, relief=RAISED)

button1.pack(side=LEFT,padx=10)
button2.pack(side="left",padx=10)
button3.pack(padx=10)


Button(root, text="Reset", font=("Calibri",20,"bold"), bg="SteelBlue1", command=reset_game).pack(pady=20)


root.mainloop()
    

```

```python
import tkinter as tk 
from tkinter import Entry, Label, Scale, Frame


class RenkPaleti:
    
    def __init__(self,gui):
        self.gui=gui
        self.gui.title("Renk Paleti")
        self.gui.geometry("300x450")
        self.gui.resizable(0,0)
        self.hex=''
        
        
        self.color_canvas=Label(self.gui ,bg='#8400ff', width=40, height=10, bd=2, relief=RAISED)
        #self.color_canvas.bind("<Double-1>",self.copyColor)
        self.color_canvas.place(x=7,y=2)
        
        
        frame=Frame(self.gui,bd=2,relief=SUNKEN)
        frame.place(x=7,y=170, height=170, width=285)
        
        red_label=Label(frame, text="R", font=("Calibri",10,"bold"), fg="red")#fg="#ff0000"
        red_label.place(x=5,y=20)
        self.Red_Scale=Scale(frame,from_=0, to=255, length=210, fg="red", orient=HORIZONTAL, command=self.scrollMove) 
        self.Red_Scale.place(x=40,y=0)
        
        green_label=Label(frame,text="G",font=("Calibri",10,"bold"),fg="green") #fg="#00a200"
        green_label.place(x=5,y=68)
        self.Green_Scale=Scale(frame, from_=0, to=255, length=210, fg="green", orient=HORIZONTAL, command=self.scrollMove)
        self.Green_Scale.place(x=40,y=50)
        
        blue_label=Label(frame,text="B",font=('Calibri',10,'bold'),fg="blue") #fg="#0000ff"
        blue_label.place(x=5,y=122)
        self.Blue_Scale=Scale(frame, from_=0, to=255, length=210, fg="blue", orient=HORIZONTAL, command=self.scrollMove)
        self.Blue_Scale.place(x=40,y=105)
        
        hex_label=Label(self.gui, text='Hex Kodu:', font=("Calibri",10,'bold'))
        hex_label.place(x=7,y=360)
        self.hex_entry=Entry(self.gui,width=12, font=("Calibri",10))
        self.hex_entry.place(x=90,y=360)
        
        rgb_label=Label(self.gui, text="RGB Kodu: ", font=("Calibri",10,'bold'))
        rgb_label.place(x=7,y=390)
        
        self.rgb_entry=Entry(self.gui, width=12, font=("Calibri",10))
        self.rgb_entry.place(x=90,y=390)
        
        
    def scrollMove(self, *args):
        r=int(self.Red_Scale.get())
        g=int(self.Green_Scale.get())
        b=int(self.Blue_Scale.get())
            
        rgb=f"{r},{g},{b}"
            
        #Convert RGB to HEX...
        self.hex="#%02x%02x%02x"%(r,g,b)

        self.color_canvas.config(bg=self.hex)
            
        self.hex_entry.delete(0,END)
        self.hex_entry.insert(0,self.hex)
            
        self.rgb_entry.delete(0,END)
        self.rgb_entry.insert(0,rgb)
            


root=tk.Tk()
RenkPaleti(root)
root.mainloop()
```

```python

###### Modüllerin dahil edilmesi###########################

import random
from tkinter import * 
import tkinter.messagebox as box

############ Global değişkenlerin tanımlanması #####################

sayi1=sayi2=toplam=0

dogru=yanlis=0
hak=5

#################### fonksiyonların tanımlanması #####################

def rastgele_sayi():
    
    global sayi1,sayi2,toplam
    
    sayi1=random.randint(100,500)
    sayi2=random.randint(100,500)
    sayi1_label.configure(text=str(sayi1))
    sayi2_label.configure(text=str(sayi2))
    
    toplam=sayi1+sayi2


def toplam_sonuc_karsilastirma():
    
    global yanlis, dogru, toplam, hak
    
    
    sonuc=int(sonuc_alani.get())

    if(toplam==sonuc):
        dogru+=1
        dogru_label_sayi.configure(text=str(dogru))
        toplam=0
        rastgele_sayi()
        
        if(dogru%5==0):
            hak+=1
            hak_label_sayi.configure(text=str(hak))
        
        
    elif(toplam!=sonuc):
        yanlis+=1
        yanlış_label_sayi.configure(text=str(yanlis))
        toplam=0
        rastgele_sayi()
        
        hak-=1
        hak_label_sayi.configure(text=str(hak))
        
        if(hak==0):
            kontrol_btn.configure(state="disabled")
            mesaj= "Hakkınız Bitti! \nDogru:{}  Yanlış:{}".format(dogru,yanlis)
            box.showinfo('Bilgi',mesaj)
            

###### Pencere Tanımı ##########################

gui=Tk()
gui.title("Toplama Oyunu")
gui.geometry("380x500")
gui.resizable(0,0)

######## Buton, Etiket, Girdi Alanı Tanımı ###################################################################

kontrol_btn=Button(gui, text="Kontrol",font=("Calibri",22),bg="coral",command=toplam_sonuc_karsilastirma)
kontrol_btn.place(x=135,y=270)

sayi1_label=Label(gui,text="0",font=("Calibri",30))
sayi1_label.place(x=160,y=50)

sayi3_label=Label(gui,text="+_____________",font=("Calibri",30))
sayi3_label.place(x=40,y=140)

sayi2_label=Label(gui,text="0",font=("Calibri",30))
sayi2_label.place(x=160,y=120)

sonuc_alani=Entry(gui,font=("Calibri",20))
sonuc_alani.insert(END,0)
sonuc_alani.place(x=50,y=200)


#### Alt Bilgilendirme Widget ##########################################

dogru_label_sayi=Label(gui,text="-",font=("Calibri",30))
dogru_label_sayi.place(x=65,y=400)

dogru_label=Label(gui,text="Doğru",font=("Calibri",20))
dogru_label.place(x=40,y=370)

yanlış_label_sayi=Label(gui,text="-",font=("Calibri",30))
yanlış_label_sayi.place(x=155,y=400)

yanlış_label=Label(gui,text="Yanlış",font=("Calibri",20))
yanlış_label.place(x=135,y=370)

hak_label_sayi=Label(gui,text="-",font=("Calibri",30))
hak_label_sayi.place(x=270,y=400)

hak_label=Label(gui,text="Kalan Hak",font=("Calibri",20))
hak_label.place(x=230,y=370)

###################################################################

rastgele_sayi()

hak_label_sayi.configure(text=str(hak))

gui.mainloop()
```
