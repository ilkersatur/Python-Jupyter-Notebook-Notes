# Papagan Sınıfı Olusturma

```python
class Papagan:
    
    #class attribute
    tür='kus'
    
    def __init__(self,ad,yas):
        self.ad=ad
        self.yas=yas
```

```python
mavis=Papagan('Mavi',5) # Papagan sınıfından mavis isimli nesne olusturalım
geveze=Papagan('Geveze',3)# Papagan sınıfından geveze isimli nesne olusturalım
```

```python
# sınıf özelligi ulasma 
print(Papagan.tür)
print(mavis.__class__.tür)
print(geveze.__class__.tür)
```

```python
# nesne özelliklerine ulasma
print("{} papagani {} yasindadir".format(mavis.ad,mavis.yas))
print("{} papagani {} yasindadir".format(geveze.ad,geveze.yas))
```

# Sınıf fonksiyonları ekleyelim

```python
class Papagan1:
    
    #class attribute
    tür='kus'
    
    def __init__(self,ad,yas):
        self.ad=ad
        self.yas=yas
    
    # nesne fonksiyonu
    def konusma(self, kelime):
        return ("{} konusur {}".format(self.ad, kelime))
    
    def dans(self):
        return ("{} dans ediyor..".format(self.ad))
    
```

```python
yesil=Papagan1('Yesil',5)
konuskan=Papagan1('Konuskan',3)
```

```python
# nesne özelliklerine ulasma
print("{} papagani {} yasindadir".format(yesil.ad,yesil.yas))
print("{} papagani {} yasindadir".format(konuskan.ad,konuskan.yas))
```

```python
yesil.dans()
```

# Inheritance (Kalıtım)

```python
class Kus:
    
    def __init__(self):
        print("Kus ortak isimdir")
    
    def buNedir(self):
        print("Kus sınıfı")     
    
    def yumurtlama(self):
        print("kuslar kuluckaya yatar")
  
```

```python
class Kartal(Kus):
    pass
```

```python
kartal1=Kartal()
```

```python
kartal1.buNedir()
```

```python
kartal1.yumurtlama()
```

```python
class Penguen(Kus):
    pass
```

```python
penguen1=Penguen()
```

```python
penguen1.buNedir()
```

```python
penguen1.yumurtlama()
```

### self kullanılmadığında hata alırız.

```python
class Kus_Sınıf:
    
    def __init__(self):
        print("Kus ortak isimdir")
    
    def buNedir(): # self kullanmadık
        print("Kus sınıfı")     
    
    def yumurtlama(self):
        print("kuslar kuluckaya yatar")
```

```python
kus=Kus_Sınıf()
```

```python
kus.buNedir() # Kus_sınıfı icerisinde buNedir() sınıfına self anahtar kelimesini eklemedik. Hata verecektir.
```

```python
kus.yumurtlama() # Kus_sınıfı icerisinde yumurtlama() sınıfına self anahtar kelimesini ekledik. Hata verecektir.
```
