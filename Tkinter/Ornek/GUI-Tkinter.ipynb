### Geometri Yöneticisi (Geometry Manager)

### The three geometry managers are as follows:

### • pack: Daha basit düzenler için kullanımı basittir, ancak biraz karmaşık düzenler için çok karmaşık olabilir.

### • grid: Bu, kolay yerleşim yönetimi için tablo benzeri bir yönetim özellikleri düzeni sağlayan en yaygın kullanılan geometri yöneticisidir.

### • place: Bu en az popüler olanıdır, ancak widget'ların mutlak konumlandırması için en iyi kontrolü sağlar.

## pack() parametreleri


#### Pakette en sık kullanılan parametreler şunları içerir:

#### side: LEFT, TOP, RIGHT ve BOTTOM (bunlar, widget'ın hizalamasına karar verir)

#### fill:X, Y, BOTH ve NONE (bunlar widget'ın boyutunun büyüyüp büyüyemeyeceğini belirler)

#### expand: Boolean values such as tkinter.YES/tkinter.NO, 1/0, True/ False

#### anchor: NW, N, NE, E, SE, S, SW, W ve CENTER (ana yönlere karşılık gelir)

#### Internal padding (ipadx and ipady) for the padding inside widgets and external padding (padx and pady), which all default to a value of zero

#### Widget'ların içindeki boşluk için ipad (ipadx ve ipady) kullanılır. İlk değerleri sıfırdır.(padx=0 ve pady=0)

####  Widget'ların içindeki dışındaki için pad (padx ve pady) kullanılır. İlk değerleri sıfırdır.(ipadx=0 ve ipady=0)



```python
from tkinter import *

frame=Tk()
frame.geometry("300x300")


Label(frame,text="side ve fill parametreleri",font=("Calibri",20)).pack()

Button(frame, text="A",font=("Calibri",20)).pack(side=TOP)
Button(frame, text="B",font=("Calibri",20)).pack(side=TOP)
Button(frame, text="C",font=("Calibri",20)).pack(side=TOP)
Button(frame, text="D",font=("Calibri",20)).pack(side=TOP)

frame.mainloop()
```

### Sticky

#### Hücre, widget'tan daha büyükse, sticky seçeneği, widget'in hangi tarafa yapışması gerektiğini ve hücre içinde, widget tarafından orijinal boyutunda alınmayan fazladan boşluğun nasıl dağıtılacağını belirtir.




```python
from tkinter import *
import calendar

frame=Tk()
frame.geometry("300x300")

Label(frame,text="Name",font=("Calibri",20)).grid(row=0,column=0,sticky=E)
Label(frame,text="Password",font=("Calibri",20)).grid(row=1,column=0,sticky=W)

Entry(frame).grid(row=0, column=1,sticky=E)
Entry(frame).grid(row=1, column=1,sticky=E)

Button(frame,text="Giris Yap",bg="aqua").grid(row=100,column=1,sticky=E)

frame.mainloop()
```

```python
from tkinter import *
import calendar

frame=Tk()
frame.geometry("300x300")

Label(frame,text="Name").grid(row=0,column=0,sticky=NW,ipadx=50,ipady=50)
Label(frame,text="Password",font=("Calibri",20)).grid(row=1,column=0,sticky=N)

Entry(frame).grid(row=0, column=1,sticky=E)
Entry(frame).grid(row=1, column=1,sticky=E)

Button(frame,text="Giris Yap",bg="aqua").grid(row=100,column=1,sticky=E)

frame.mainloop()
```
