```python


def topla():
    global toplam
    
    sayi1=int(sayi1_girdi.get())
    sayi2=int(sayi2_girdi.get())
    toplam= sayi1+sayi2 
    
    sonuc_alani.insert(1, str(toplam))
    
    
def kaydet_toplama():
    
    dosya=open("sonuclar.txt","a")
    bilgi="Toplam Sonucu:{}\n".format(toplam)
    dosya.write(bilgi)
    dosya.close()
    
    
def carpma():
    global carpma
    
    sayi1=int(sayi1_girdi.get())
    sayi2=int(sayi2_girdi.get())
    carpma= sayi1*sayi2 
    
    sonuc_alani.insert(1, str(carpma))
    
def kaydet_carpma():
    
    dosya=open("sonuclar.txt","a")
    bilgi="Carpma Sonucu:{}\n".format(carpma)
    dosya.write(bilgi)
    dosya.close()


def alt_pencere1():
    
    global sayi1_girdi,sayi2_girdi,sonuc_alani
    
    gui1=Toplevel(root)
    
    gui1.geometry("350x350")
    gui1.resizable(0,0)
    gui1.title("Alt Pencere-1")
    gui1.config(bg="palegreen1")
    
    sayi1_label=Label(gui1,text="Sayi1:",font=("Calibri",15),bg="palegreen1")
    sayi1_label.place(x=40,y=40)
    sayi2_label=Label(gui1,text="Sayi2:",font=("Calibri",15),bg="palegreen1")
    sayi2_label.place(x=40,y=100)
    
    
    sayi1_girdi=Entry(gui1,font=("Calibri",15))
    sayi1_girdi.place(x=100,y=40)
    sayi2_girdi=Entry(gui1,font=("Calibri",15))
    sayi2_girdi.place(x=100,y=100)
    
    sonuc_label=Label(gui1,text="Sonuc:",font=("Calibri",15),bg="palegreen1")
    sonuc_label.place(x=40,y=160)
    
    sonuc_alani=Entry(gui1,font=("Calibri",15))
    sonuc_alani.place(x=100,y=160)
    
    toplam_btn=Button(gui1, text="Topla",width=6, font=("Calibri",20,"bold"),command=topla)
    toplam_btn.place(x=70,y=220)
    
    
    kaydet_btn=Button(gui1, text="Kaydet",width=6, font=("Calibri",20,"bold"),command=kaydet_toplama)
    kaydet_btn.place(x=200,y=220)
    
    gui1.mainloop()

    
def alt_pencere2(event):
    
    global sayi1_girdi,sayi2_girdi,sonuc_alani
    
    gui2=Toplevel(root)
    
    gui2.geometry("350x350")
    gui2.resizable(0,0)
    gui2.title("Alt Pencere-2")
    gui2.config(bg="light blue")
    

    
    sayi1_label=Label(gui2,text="Sayi1:",font=("Calibri",15),bg="light blue")
    sayi1_label.place(x=40,y=40)
    sayi2_label=Label(gui2,text="Sayi2:",font=("Calibri",15),bg="light blue")
    sayi2_label.place(x=40,y=100)
    
    
    sayi1_girdi=Entry(gui2,font=("Calibri",15))
    sayi1_girdi.place(x=100,y=40)
    sayi2_girdi=Entry(gui2,font=("Calibri",15))
    sayi2_girdi.place(x=100,y=100)
    
    sonuc_label=Label(gui2,text="Sonuc:",font=("Calibri",15),bg="light blue")
    sonuc_label.place(x=40,y=160)
    
    sonuc_alani=Entry(gui2,font=("Calibri",15))
    sonuc_alani.place(x=100,y=160)
    
    carpma_btn=Button(gui2, text="Carpma",width=6, font=("Calibri",20,"bold"),command=carpma)
    carpma_btn.place(x=70,y=220)
    
    kaydet_btn=Button(gui2, text="Kaydet",width=6, font=("Calibri",20,"bold"),command=kaydet_carpma)
    kaydet_btn.place(x=200,y=220)
    
    gui2.mainloop()


from tkinter import * 

root=Tk()

root.geometry("300x300")
root.resizable(0,0)
root.title("Ana Pencere")

Button(root, text="ISLEM-1", font=("Calibri",20,"bold"),bd=4, bg="seagreen1",command=alt_pencere1).pack(pady=50)
btn=Button(root, text="ISLEM-2", font=("Calibri",20,"bold"),bd=4, bg="steelblue1")
btn.pack()
btn.bind("<Button-1>",alt_pencere2)


root.mainloop()
```
