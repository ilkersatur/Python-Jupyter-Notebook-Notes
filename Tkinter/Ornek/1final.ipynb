```python
import random
ds=0
ys=0
beraberes=0
haks=10
bilgisayar_deger={
    "0":"Tas",
    "1":"Kagit",
    "2":"Makas"}

def pencere2():
    def Tas():
        global beraberes

        tahmin=bilgisayar_deger[str(random.randint(0,2))]
        pc.configure(text=tahmin)
        if(tahmin=="Tas"):
            karar.configure(text="berabere")
            beraberes+=1
            berabere.configure(text=beraberes)
        elif(tahmin=="Kagit"):
            karar.configure(text="win")
            
        else:
            karar.configure(text="lose")
    def Kagit():
        tahmin=bilgisayar_deger[str(random.randint(0,2))]
        pc.configure(text=tahmin)
        if(tahmin=="Kagit"):
            karar.configure(text="berabere")
        elif(tahmin=="Tas"):
            karar.configure(text="lose")
        else:
            karar.configure(text="win")
    def Makas():
        tahmin=bilgisayar_deger[str(random.randint(0,2))]
        pc.configure(text=tahmin)
        if(tahmin=="Makas"):
            karar.configure(text="berabere")
        elif(tahmin=="Tas"):
            karar.configure(text="lose")
        else:
            karar.configure(text="win")
            
    oyun=Toplevel(ana)
    oyun.geometry("300x300")
    
    karar=Label(oyun,text="sonuc")
    karar.place(x=30,y=40)
   
    pc=Label(oyun,text="PC")
    pc.place(x=10,y=10)
    
    dogru=Label(oyun,text="doğru")
    dogru.place(x=10,y=120)
    yanlis=Label(oyun,text="yanlış")
    yanlis.place(x=10,y=150)
    berabere=Label(oyun,text="berabere")
    berabere.place(x=10,y=200)
    hak=Label(oyun,text="hak")
    hak.place(x=10,y=250)
    
    button1=Button(oyun,text="TAS",  font=("Calibri",5,"bold"), command=Tas, relief=RAISED)
    button2=Button(oyun,text="KAGIT",  font=("Calibri",5,"bold"), command=Kagit, relief=RAISED)
    button3=Button(oyun,text="MAKAS",  font=("Calibri",5,"bold"), command=Makas, relief=RAISED)

    button1.place(x=30,y=80)
    button2.place(x=30,y=100)
    button3.place(x=30,y=60)
    
def pencere1():
    
    def kaydet(e):
        dosya1 = open('sonuc.txt', 'w')
        satir1 = "Sonuc={}".format(toplam)
        dosya1.write(satir1) 
        dosya1.close()
    def sifirla(e):
        sonuc.configure(text="")
        hesapla.config(state="active")
    def topla(e):
        global toplam
        sayi1=int(sayi1_girdi.get())
        sayi2=int(sayi2_girdi.get())    
        toplam=sayi1+sayi2
        sonuc.configure(text=toplam)
        hesapla.config(state="disable")
    def cıkar(e):
        sayi1=int(sayi1_girdi.get())
        sayi2=int(sayi2_girdi.get())    
        toplam=sayi1-sayi2
        sonuc.configure(text=toplam)
    def bol(e):
        sayi1=int(sayi1_girdi.get())
        sayi2=int(sayi2_girdi.get())    
        toplam=sayi1/sayi2
        sonuc.configure(text=toplam)
    def carp(e):
        sayi1=int(sayi1_girdi.get())
        sayi2=int(sayi2_girdi.get())    
        toplam=sayi1*sayi2
        sonuc.configure(text=toplam)
        
    hesap=Toplevel(ana)
    hesap.geometry("300x300")
    
    sayi1_girdi=Entry(hesap,font=("Arial",5))
    sayi1_girdi.place(x=40,y=50)
    sayi2_girdi=Entry(hesap,font=("Arial",5))
    sayi2_girdi.place(x=40,y=80)
    info1=Label(hesap,text="sayi gir")
    info1.place(x=40,y=10)
    hesapla=Button(hesap,text="sağ tık topla, sol tık çıkar, orta tık çarp, enter böl, boş enter herşeyi sıfırla")
    hesapla.bind('<Button-1>',topla)
    hesapla.bind('<Button-2>',cıkar)
    hesapla.bind('<Button-3>',bol)
    hesapla.bind('<Return>',carp)
    hesapla.focus()
    hesapla.place(x=0,y=100)
    sonuc=Label(hesap,text="sonuc")
    sonuc.place(x=40,y=130)
    hesap.bind('<Return>',kaydet)
    sifir=Button(hesap)
    sifir.place(x=40,y=160)
    sifir.bind('<Button>',sifirla)
from tkinter import*
ana=Tk()
ana.geometry("300x300")
birinci=Button(ana,text="Hesap",font=("Arial",20),relief=GROOVE,command=pencere1)
birinci.focus()
birinci.place(x=100,y=50)
ikinci=Button(ana,text="oyun",font=("Arial",20),relief=GROOVE,command=pencere2)
ikinci.focus()
ikinci.place(x=100,y=150)
ana.mainloop()
```
