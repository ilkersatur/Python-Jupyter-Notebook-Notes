```python
import math

def math_PI():
    
    sonuc_text.insert(1,str(math.pi))
    
def toplama_fcn():
    sayi1= int(sayi1_text.get())
    sayi2= int(sayi2_text.get())
    
    toplam=sayi1+sayi2
    sonuc_text.insert(1,str(toplam))

def cikarma_fcn():
    sayi1= int(sayi1_text.get())
    sayi2= int(sayi2_text.get())
    
    cikarma=sayi1-sayi2
    sonuc_text.insert(1,str(cikarma))

def carpma_fcn():
    sayi1= int(sayi1_text.get())
    sayi2= int(sayi2_text.get())
    
    carpma=sayi1*sayi2
    sonuc_text.insert(1,str(carpma))


def us_alma_fcn():
    
    sayi1= int(sayi1_text.get())
    sayi2= int(sayi2_text.get())
    
    power=sayi1**sayi2
    sonuc_text.insert(1,str(power))
    


def temizle_fcn():
    sayi1_text.delete(0,END)
    sayi2_text.delete(0,END)
    sonuc_text.delete(0,END)




from tkinter import *

gui=Tk()

gui.configure(bg="light yellow")
gui.title("Hesap Makinesi")
gui.geometry("500x500")


# pencere içerisinde yer alacak pencere ögeleri (widget)

sayi1_etiket=Label(gui, text="Sayi1", bg="light green")
sayi1_etiket.grid(row=1,column=1, padx=10,pady=10)


sayi2_etiket=Label(gui, text="Sayi2", bg="light green")
sayi2_etiket.grid(row=2,column=1, padx=10,pady=10)


sonuc_etiket=Label(gui, text="Sonuc", bg="light green")
sonuc_etiket.grid(row=4,column=1, padx=10,pady=10)


topla_btn=Button(gui, text="+", fg="black", bg="salmon", font=("Calibri",20), command=toplama_fcn)
topla_btn.grid(row=6, column=0, padx=10, pady=10,ipady=10)


cikarma_btn=Button(gui, text="-", fg="black", bg="salmon", font=("Calibri",20), command=cikarma_fcn)
cikarma_btn.grid(row=6, column=1, padx=10, pady=10,ipady=10)


carpma_btn=Button(gui, text="x", fg="black", bg="salmon", font=("Calibri",20), command=carpma_fcn)
carpma_btn.grid(row=7, column=0, padx=10, pady=10,ipady=10)


us_alma_btn=Button(gui, text="pow", fg="black", bg="salmon", font=("Calibri",20), command=us_alma_fcn)
us_alma_btn.grid(row=7, column=1, padx=10, pady=10,ipady=10)

pi_btn=Button(gui, text="Pi", fg="black", bg="salmon", font=("Calibri",20), command=math_PI)
pi_btn.grid(row=6, column=2, padx=10, pady=10,ipady=10)

temizle_btn=Button(gui, text="Clear", fg="black", bg="salmon", font=("Calibri",20), command=temizle_fcn)
temizle_btn.grid(row=6, column=3, padx=10, pady=10,ipady=10)




sayi1_text=Entry(gui)
sayi2_text=Entry(gui)
sonuc_text=Entry(gui)


sayi1_text.grid(row=1,column=2, padx=10, pady=10)
sayi2_text.grid(row=2,column=2, padx=10, pady=10)
sonuc_text.grid(row=4,column=2, padx=10, pady=10)

gui.mainloop()
```
