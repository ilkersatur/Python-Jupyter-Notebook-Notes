```python
import random
rastgelesekil={
    "0":"Altıgen",
    "1":"Beşgen",
    "2":"Dörtgen",
    "3":"Sekizgen",
    "4":"Üçgen",
    "5":"Yedigen",}

def uretim():
    
    rsekil=rastgelesekil[str(random.randint(0,5))]
    
    if(rsekil=="Dörtgen"):
        sekil.configure(text="Dörtgen")
        icacisi.configure(text="90")        
        disacisi.configure(text="270")
        kenarsayisi.configure(text="4")
        
        dortgen=PhotoImage(file="Dörtgen.png")
        frame=Label(ana,image=dortgen)
        frame.image=dortgen
        frame.place(x=100,y=30)
        
    if(rsekil=="Altıgen"):
        sekil.configure(text="Altıgen")
        icacisi.configure(text="120")        
        disacisi.configure(text="60")
        kenarsayisi.configure(text="6")
        
        altıgen=PhotoImage(file="Altıgen.png")
        frame=Label(ana,image=altıgen)
        frame.image=altıgen
        frame.place(x=100,y=30)
        
    if(rsekil=="Beşgen"):
        sekil.configure(text="Beşgen")
        icacisi.configure(text="108")        
        disacisi.configure(text="72")
        kenarsayisi.configure(text="5")
        
        besgen=PhotoImage(file="Beşgen.png")
        frame=Label(ana,image=besgen)
        frame.image=besgen
        frame.place(x=100,y=30)
    if(rsekil=="Sekizgen"):
        sekil.configure(text="Sekizgen")
        icacisi.configure(text="135")        
        disacisi.configure(text="45")
        kenarsayisi.configure(text="8")
        
        sekizgen=PhotoImage(file="Sekizgen.png")
        frame=Label(ana,image=sekizgen)
        frame.image=sekizgen
        frame.place(x=100,y=30)
    if(rsekil=="Üçgen"):
        sekil.configure(text="Üçgen")
        icacisi.configure(text="60")        
        disacisi.configure(text="120")
        kenarsayisi.configure(text="3")
        
        ucgen=PhotoImage(file="Üçgen.png")
        frame=Label(ana,image=ucgen)
        frame.image=ucgen
        frame.place(x=100,y=30)
    if(rsekil=="Yedigen"):
        sekil.configure(text="Yedigen")
        icacisi.configure(text="129")        
        disacisi.configure(text="51")
        kenarsayisi.configure(text="7")
        
        yedigen=PhotoImage(file="Yedigen.png")
        frame=Label(ana,image=yedigen)
        frame.image=yedigen
        frame.place(x=100,y=30)

        
from tkinter import *
ana=Tk()
ana.geometry("400x600")

sekil=Label(ana, text="   ",font=("Arial",10))
kenarsayisi_text=Label(ana, text="Kenar Sayısı",font=("Arial",10))
disacisi_text=Label(ana, text="Bir Dış Açısı",font=("Arial",10))
icacisi_text=Label(ana, text="Bir iç Açısı",font=("Arial",10))
disacisi=Label(ana, text="   ",font=("Arial",10))
kenarsayisi=Label(ana, text="  ",font=("Arial",10))
icacisi=Label(ana, text=" ",font=("Arial",10))

sekil.place(x=100,y=290)
kenarsayisi_text.place(x=100,y=320)
icacisi_text.place(x=100,y=350)
disacisi_text.place(x=100,y=380)

kenarsayisi.place(x=300,y=320)
icacisi.place(x=300,y=350)
disacisi.place(x=300,y=380)

uret=Button(ana,text="ÜRET",  font=("Calibri",20,"bold"), relief=RAISED,command=uretim)
uret.place(x=100,y=440)

ana.mainloop()
```
