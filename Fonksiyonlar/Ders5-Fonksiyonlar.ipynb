# 3-Bir kullanıcıdan harf girişi isteyelim. 
### * Girilen harf H veya h’a eşitse kullanıcıdan üçgenin iki kenar bilgisini ana fonksiyonda alarak hipotenusunu hesaplayıp fonksiyonda yazdıralım.
### * Girilen harf U veya u’ya eşitse kullanıcıdan üçgenin kenar bilgilerini alt fonksiyonda çevresini hesaplayıp ana fonksiyona döndürelim. 
### * Girilen harf D veya d ’ya eşitsekullanıcıdan yarıcap bilgisini alt fonksiyonda alarak dairenin alanını ve çevresini hesaplayalım ve ana fonksiyona döndürelim buradan yaz() fonksiyonuna göndererek sonucu yazdıralım
### * Girilen harf R veya R  ise kullanıcıdan dikdörtgenin kenar bilgilerini alarak alanını ve çevresini hesaplayarak sonucları yaz() adlı fonksiyona gönderelim ve sonuçları burada yazdıralım.
### * Kullanıcı Q veya q harfi girdiğinde program sonlansın.


```python
def hipotenus(gelen_kenar1, gelen_kenar2):
    
    sonuc= ((gelen_kenar1**2)+(gelen_kenar2**2))**0.5
    
    print("Hipotenus:",sonuc)
```

```python
def cevre_ucgen():
    
    kenar1=float(input("Birinci kenari giriniz:"))
    kenar2=float(input("İkinci kenari giriniz:"))
    kenar3=float(input("Ücüncü kenari giriniz:"))
    
    cevre=kenar1+kenar2+kenar3
    
    return(cevre)
    
```

```python
def daire_bilgi():
    
    yaricap=float(input("Birinci kenari giriniz:"))
    
    alan=3.14*yaricap*yaricap
    cevre= 2*3.14*yaricap
    
    return(cevre, alan)
    
```

```python
def yaz(gelen_bilgi1, gelen_bilgi2):
    
    print("Dairenin Cevresi:",gelen_bilgi1)
    print("Dairenin Alani:",gelen_bilgi2)
```

```python

while(True):
    # Ana fonksiyon
    print("""
    H veya h ise Ücgen Hipotenus hesabı
    U veya u ise Ücgen Alan ve Cevre
    D veya d ise Dairenin Alanı ve Cevre
    R veya r ise Dikdortgen Alanı ve Cevre
    Q veya q ise Cıkıs
    """)

    harf=input("Bir harf giriniz:")

    if(harf=='H' or harf== 'h'):

        kenar1=float(input("Birinci kenari giriniz:"))
        kenar2=float(input("İkinci kenari giriniz:"))

        hipotenus(kenar1,kenar2)


    if(harf=='U' or harf== 'u'):

        sonuc=cevre_ucgen()
        print("Ucgenin cevresi",sonuc)


    if(harf=='D' or harf== 'd'):

        daire_cevre,daire_alan=daire_bilgi()
        yaz(daire_cevre,daire_alan)
    

# if(harf=='R' or harf== 'r') bu kısmı siz yapın lütfen!

#Girilen harf R veya R  ise kullanıcıdan dikdörtgenin kenar bilgilerini alarak alanını ve çevresini hesaplayarak sonucları 
# yaz() adlı fonksiyona gönderelim ve sonuçları burada yazdıralım.
    
   
    if( harf=="Q" or harf=="q"):
        break
        
    
    else:
        print("Yanlıs deger girdiniz!Lütfen dogru harfi giriniz!")
```
